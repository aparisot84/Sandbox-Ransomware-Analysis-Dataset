```python
filepath = "..//7 - TF-IDF"
filelist = os.listdir(filepath)
filelist.sort()
```

```python
for file in filelist:
    print(file)
    with open(filepath + "//" + file, "r") as data:
        content = data.readlines()
```

```python
def load_files():
    for file in filelist:
        print(file)
        with open(filepath + "//" + file, "r") as data:
            conteudo = data.readlines()

    return conteudo
```

```python
vec=TfidfVectorizer()
matrix=vec.fit_transform(load_files())
```

```python
print("Feature Names n",vec.get_feature_names_out())
len(vec.get_feature_names_out())
```

```python
print("Sparse Matrix n",matrix.shape,"n",matrix.toarray())
```
